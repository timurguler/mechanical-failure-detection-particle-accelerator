```python
import datetime
import pandas as pd
from os import listdir
from os.path import isfile, join
import matplotlib
import matplotlib.pyplot as plt
from IPython.display import IFrame
import plotly.graph_objects as go
import plotly.express as px
```

```python




##### UPDATE! ######

### will need all the faults in one folder and the runs in a seperate folder ###

mypath = '/Users/colinobrien/Desktop/Repo/capstone/10_13_21/RFQ_UVA_Capstone'
mypath_runs = '/Users/colinobrien/Desktop/Repo/capstone/10_13_21/RFQ_UVA_Capstone/runs'
mypath_faults= '/Users/colinobrien/Desktop/Repo/capstone/10_13_21/RFQ_UVA_Capstone/faults'
```

```python
## getting all the files:
all_files = [f for f in listdir(mypath) if isfile(join(mypath, f))]
all_runs = [f for f in listdir(mypath_runs) if isfile(join(mypath_runs, f))]
all_faults = [f for f in listdir(mypath_faults) if isfile(join(mypath_faults, f))]
```

```python
## replacing pulling out just the date and time:

#faults:
all_faults_df = pd.DataFrame(all_faults)
all_faults_df = all_faults_df.replace('RFQ_400ns_Fault_', '', regex=True)
all_faults_df = all_faults_df.replace('.csv', '', regex=True)
all_faults_df = all_faults_df.replace('.xlsx', '', regex=True)
all_faults_df = all_faults_df.replace('.xls', '', regex=True)

# runs
#faults:
all_runs_df = pd.DataFrame(all_runs)
all_runs_df = all_runs_df.replace('RFQ_400ns_Run_', '', regex=True)
all_runs_df = all_runs_df.replace('.csv', '', regex=True)
all_runs_df = all_runs_df.replace('.xlsx', '', regex=True)
all_runs_df = all_runs_df.replace('.xls', '', regex=True)
```

```python
## converting to datetime:
## faults
faults_date_time = []
for i in all_faults_df[0]:
    date_time_obj = datetime.datetime.strptime(i, '%m.%d.%Y-%I.%M.%S%p')
    faults_date_time.append(date_time_obj)
faults_date_time.sort()
faults_date_time_df2 = pd.DataFrame(faults_date_time)

## runs
runs_date_time = []
for i in all_runs_df[0]:
    date_time_obj = datetime.datetime.strptime(i, '%m.%d.%Y-%I.%M.%S%p')
    runs_date_time.append(date_time_obj)
runs_date_time.sort()
runs_date_time_df2 = pd.DataFrame(runs_date_time)


```

```python
## breaking out columns of the DataFrame:
## faults:
year_faults = []
for i in faults_date_time:
    local_year = i.year
    year_faults.append(local_year)
    
month_faults = []
for i in faults_date_time:
    local_month = i.month
    month_faults.append(local_month)
    
day_faults = []
for i in faults_date_time:
    local_day = i.day
    day_faults.append(local_day)
    
hour_faults = []
for i in faults_date_time:
    local_hour = i.hour
    hour_faults.append(local_hour)

        
min_faults = []
for i in faults_date_time:
    local_min = i.minute
    min_faults.append(local_min)
    
sec_faults = []
for i in faults_date_time:
    local_sec = i.second
    sec_faults.append(local_sec)
    
    
faults_df = pd.DataFrame({'year':year_faults,
                         'month':month_faults,
                         'day':day_faults,
                         'hour':hour_faults,
                         'min':min_faults,
                         'sec':sec_faults})


## runs:
year_runs = []
for i in runs_date_time:
    local_year = i.year
    year_runs.append(local_year)
    
month_runs = []
for i in runs_date_time:
    local_month = i.month
    month_runs.append(local_month)
    
day_runs = []
for i in runs_date_time:
    local_day = i.day
    day_runs.append(local_day)
    
hour_runs = []
for i in runs_date_time:
    local_hour = i.hour
    hour_runs.append(local_hour)

        
min_runs = []
for i in runs_date_time:
    local_min = i.minute
    min_runs.append(local_min)
    
sec_runs = []
for i in runs_date_time:
    local_sec = i.second
    sec_runs.append(local_sec)
    
    
runs_df = pd.DataFrame({'year':year_runs,
                         'month':month_runs,
                         'day':day_runs,
                         'hour':hour_runs,
                         'min':min_runs,
                         'sec':sec_runs})
```

```python
## re-arranging data for plots:

## faults:

## grouping by the year, month, day:
grouped_faults = faults_df.groupby(['year','month', 'day']).size().reset_index(name='counts')

## need to get the x axis in datetime back to datetime:
just_month_and_day_fault = grouped_faults[['year','month', 'day']]
month_day_datetime_fault = []
for index, row in just_month_and_day_fault.iterrows():
    just_year = str(row['year'])
    just_month = str(row['month'])
    just_day = str(row['day'])
    month_day = just_year+'-'+just_month+'-'+just_day
    local_dt_obj = datetime.datetime.strptime(month_day, '%Y-%m-%d')
    month_day_datetime_fault.append(local_dt_obj)
# month_day_datetime_fault


## runs:

## grouping by the year, month, day:
grouped_runs = runs_df.groupby(['year','month', 'day']).size().reset_index(name='counts')

## need to get the x axis in datetime back to datetime:
just_month_and_day_run = grouped_runs[['year','month', 'day']]
month_day_datetime_run = []
for index, row in just_month_and_day_run.iterrows():
    just_year = str(row['year'])
    just_month = str(row['month'])
    just_day = str(row['day'])
    month_day = just_year+'-'+just_month+'-'+just_day
    local_dt_obj = datetime.datetime.strptime(month_day, '%Y-%m-%d')
    month_day_datetime_run.append(local_dt_obj)
# month_day_datetime_run
```

```python
## interactive plots:

## faults:

grouped_faults['date_time'] = month_day_datetime_fault
fig = px.scatter(grouped_faults, x='date_time', y='counts', 
                 height=600, width=600,
                 labels={'date_time':'Date', 
                        'counts':'faults (Some Kind of issue) that day'},
#                  hover_data=['partyID', 'sex', 'state'],
                 title = 'Faults Data')
fig.update(layout=dict(title=dict(x=0.5)))
fig.show()


## runs:
grouped_runs['date_time'] = month_day_datetime_run
fig = px.scatter(grouped_runs, x='date_time', y='counts', 
                 height=600, width=600,
                 labels={'date_time':'Date', 
                        'counts':'Runs (No issue) that day'},
#                  hover_data=['partyID', 'sex', 'state'],
                 title = 'Runs Data')
fig.update(layout=dict(title=dict(x=0.5)))
fig.show()
```

```python
### plot of all data:
## getting df
grouped_faults['date_time'] = month_day_datetime_fault
# grouped_faults['category'] = 'fault'
grouped_faults = grouped_faults.rename(columns={'counts':'counts_faults'})

grouped_runs['date_time'] = month_day_datetime_run
# grouped_runs['category'] = 'run'
grouped_runs = grouped_runs.rename(columns={'counts':'counts_runs'})

all_runs = pd.concat([grouped_faults,grouped_runs])
all_runs = all_runs.fillna(0)
all_runs['total count'] = all_runs['counts_faults']+all_runs['counts_runs']

grouped_all_2 = all_runs.groupby('date_time').agg({'counts_faults':'max','counts_runs':'max','total count':'sum'}).reset_index()
grouped_all_2


## plot:
fig = px.scatter(grouped_all_2, x='date_time', y='total count', 
                 height=600, width=600,
                 labels={'date_time':'Date', 
                        'counts':'a run that day'},
                 hover_data=['counts_faults', 'counts_runs'],
                 title = 'All Data')
fig.update(layout=dict(title=dict(x=0.5)))
fig.show()
```
